### Data cleaning:

```python
data.isnull().sum()
```

```python
#get the unique values:

data.nunique()
```

```python
data['rating'].value_counts
```

```python
# Handling the missing values:

#rating:

data['rating'].fillna('TV-MA',inplace=True)
```

```python
data['director'].value_counts
```

```python
# Maxmium values in director column is missing, then we fill the null values as 'Unknown'.

#Director:

data['director'].fillna('Unknown',inplace=True)
```

```python
#cast

data['cast'].value_counts
```

```python
data['cast'].fillna('Unknown',inplace=True)
```

```python
data['country'].value_counts
```

```python
#Country:

data['country'].fillna(data['country'].mode()[0],inplace=True)
```

```python
data['date_added'].isnull().sum()
```

```python
data['duration'].isnull().sum()
```

```python
#date_added:

data.dropna(subset=['date_added'],inplace=True)
```

```python
data.dropna(subset=['duration'],inplace=True)
```

```python
#checking for null values:

data.isnull().sum()
```

#### All missing values has been cleared.

```python
data.dtypes
```

```python
# changing the date datatype:

data['date_added'] = pd.to_datetime(data['date_added'])
```

```python
# getting the month and year column:

data['month'] = pd.to_numeric(data['date_added'].dt.month, errors='coerce').astype('Int64')
```

```python
data['year'] = pd.to_numeric(data['date_added'].dt.year, errors='coerce').astype('Int64')
```

```python
data.head(5)
```

```python
#changing the column of Listed_on as Genre:

data = data.rename(columns = {'listed_in':'genre'})
```

```python
data['genre'] = data['genre'].apply(lambda x: x.split(",")[0])
```

```python
data.head(5)
```

### EDA:

#### 1. Which year has the maximum number of released TV shows and movies?

```python
tv_shows= data[data['type']=='TV Show']
movies= data[data['type']=='Movie']
```

```python
tv_shows['year'].value_counts()
plt.figure(figsize=(10,4))

sns.countplot(x = 'year', data = data, order = tv_shows['year'].value_counts().index)
plt.title("Most released Tv-Shows in Year")
plt.ylabel('count')
plt.xlabel('Year')
```

```python
movies['year'].value_counts()
plt.figure(figsize=(10,4))

sns.countplot(x = 'year', data = data, order = movies['year'].value_counts().index)
plt.title("Most released Movies in Year")
plt.ylabel('count')
plt.xlabel('Year')
```

#### 2. What is the distribution of content types (TV shows and movies) in the dataset?

```python
data['type'].value_counts()
```

```python
types = data['type'].value_counts()
plt.figure(figsize = (6,4))

plt.pie(types.values, labels=data['type'].value_counts().index, autopct='%1.1f%%', colors=['skyblue', 'blue'])
plt.title("Distributions of Tv shows and movies are")
plt.show()
```

#### 3. Which countries have the highest number of content releases Tv-shows and Movies?

```python
tv_shows['country'].value_counts()
plt.figure(figsize=(8,4))

sns.countplot(x='country', data=data, order = tv_shows['country'].value_counts().iloc[:10].index)
plt.xticks(rotation=90)
plt.title("Countries with highest no.of content releases")
plt.xlabel("country")
plt.ylabel("count")
```

```python
movies['country'].value_counts()
plt.figure(figsize=(8,4))

sns.countplot(x='country', data=data, order = movies['country'].value_counts().iloc[:10].index)
plt.xticks(rotation=90)
plt.title("Countries with highest no.of content releases")
plt.xlabel("Country")
plt.ylabel("Count")
```

###### - Countries with highest no.of content releases on Tv-shows and movies are United States.

#### 7. List the top 10 directors in the Netflix dataset, considering both TV shows and movies.

#### 8. How has the distribution of content releases changed over the years?

```python
counts = data['release_year'].value_counts().sort_index()
plt.figure(figsize=(10,5))

plt.plot(counts.index, counts.values, marker='*')
plt.xlabel('Release Year')
plt.ylabel('Number of Releases')
plt.title('Distribution of Content Releases Over the Years')
plt.show()
```
